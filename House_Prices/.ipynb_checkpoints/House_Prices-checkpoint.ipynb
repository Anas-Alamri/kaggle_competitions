## Data Cleaning:

```python
# The following features has NA values as in the data discription file but it is counted as Nan in Python 
# while opening of the files. We are going to convert them to the string 'NA':
# Alley,BsmtQual,BsmtCond,BsmtExposure,BsmtFinType1,BsmtFinType2,FireplaceQu,GarageType,GarageFinish,GarageQual,GarageCond,
# PoolQC,Fence,MiscFeature

features_with_na=['Alley','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','FireplaceQu',\
                  'GarageType','GarageFinish','GarageQual','GarageCond','PoolQC','Fence','MiscFeature']

for i in features_with_na:
    X[i]=X[i].map(lambda x: 'NA' if pd.isnull(x) else x)
```

####  Cleaning GarageYrBlt

```python
X.GarageYrBlt.isnull().sum()
```

```python
Gara_temp=X[['GarageYrBlt','GarageFinish','GarageType','GarageCars','GarageArea','GarageQual','GarageCond','YearBuilt']]
```

```python
Gara_temp=Gara_temp[pd.isnull(Gara_temp['GarageYrBlt'])]
```

```python
Gara_temp.shape
```

```python
Gara_temp.head()
```

GarageYrBlt with NaN value actually does not have a garage as shown in the above table. GarageYrBlt will not affect the sale price so we will not use it in our model. But to make the data headmap clean and to make sure that we did not miss any missing values, we are going to fill the empty data with the YearBuilt Value

```python
# X.GarageYrBlt=X.GarageYrBlt.map(lambda x : X['YearBuilt'] if pd.isnull(x) else x)    
# X.GarageYrBlt=X.GarageYrBlt.map(lambda x : X['YearBuilt'] if pd.isnull(x) else x)   
X['GarageYrBlt']=X['GarageYrBlt'].fillna(X['YearBuilt'])
```

```python
X.GarageYrBlt.isnull().sum()
```

####  Cleaning LotFrontage


```python
X.LotFrontage=X.LotFrontage.fillna(X.LotFrontage.mean())
```

```python
X.LotFrontage.isnull().sum()
```

####  Cleaning KitchenQual

```python
X.KitchenQual.isnull().sum()
```

```python
X.KitchenQual.value_counts()
```

```python
# We are going to take the mode for the only missing item. Mode is 'TA'

X.KitchenQual=X.KitchenQual.fillna('TA')
```

```python
X.KitchenQual.isnull().sum()
```

####  Cleaning MasVnrType

```python
X.MasVnrType.isnull().sum()
```

```python
X.MasVnrType.value_counts()
```

```python
# We are going to take the mode for the missing items. Mode is 'None'

X.MasVnrType=X.MasVnrType.fillna('None')
```

```python
X.MasVnrType.isnull().sum()
```

####  Cleaning MasVnrArea

```python
X.MasVnrArea.isnull().sum()
```

```python
# We are going to take the mean for the missing items. 

X.MasVnrArea=X.MasVnrArea.fillna(X.MasVnrArea.mean())
```

```python
X.MasVnrArea.isnull().sum()
```

####  Cleaning MSZoning

```python
X.MSZoning.isnull().sum()
```

```python
X.MSZoning.value_counts()
```

```python
# We are going to take the mode for the missing items. Mode is 'RL'

X.MSZoning=X.MSZoning.fillna('RL')
```

```python
X.MSZoning.isnull().sum()
```

```python
X.BsmtHalfBath = X.BsmtHalfBath.fillna(0.0)
X.Functional  = X.Functional.fillna('Typ')
X.Utilities  = X.Utilities.fillna('AllPub')
X.KitchenQual = X.KitchenQual.fillna('TA')
X.Exterior1st = X.Exterior1st.fillna('VinylSd')
X.Exterior2nd = X.Exterior2nd.fillna('VinylSd')
X.GarageCars = X.GarageCars.fillna(2.0)
X.GarageArea = X.GarageArea.fillna(X['GarageArea'].mean())
X.BsmtFinSF1 = X.BsmtFinSF1.fillna(X['BsmtFinSF1'].mean())
X.Electrical = X.Electrical.fillna('SBrkr')
X.SaleType = X.SaleType.fillna('WD')
X.TotalBsmtSF = X.TotalBsmtSF.fillna(X['TotalBsmtSF'].mean())
X.BsmtUnfSF = X.BsmtUnfSF.fillna(X['BsmtUnfSF'].mean())
X.BsmtFinSF2 = X.BsmtFinSF2.fillna(0.0)
X.BsmtFullBath = X.BsmtFullBath.fillna(0.0)
```
